```python
import pandas as pd
import glob
import os

# 1. Fetch all CSV files in the data folder
csv_files = glob.glob('../data/*.csv')

print(f"Found {len(csv_files)} datasets to audit.\n")

# 2. Iterate through each dataset and generate a health report
audit_summary = []

for file in sorted(csv_files):
    table_name = os.path.basename(file)
    df = pd.read_csv(file)
    
    total_rows = len(df)
    total_cols = len(df.columns)
    duplicate_rows = df.duplicated().sum()
    null_cols = df.isnull().sum().loc[lambda x: x > 0].to_dict()
    
    audit_summary.append({
        "Table Name": table_name,
        "Rows": total_rows,
        "Columns": total_cols,
        "Duplicate Rows": duplicate_rows,
        "Missing Value Columns": len(null_cols)
    })
    
    print(f"==========================================")
    print(f"TABLE: {table_name}")
    print(f"==========================================")
    print(f"Dimensions: {total_rows:,} rows × {total_cols} columns")
    print(f"Duplicates: {duplicate_rows}")
    print("Null Values Count:")
    if null_cols:
        for col, count in null_cols.items():
            pct = (count / total_rows) * 100
            print(f"  - {col}: {count:,} ({pct:.2f}%)")
    else:
        print("  - No missing values found!")
    print("\n")

# 3. Summary DataFrame
summary_df = pd.DataFrame(audit_summary)
print("=== AUDIT SUMMARY OVERVIEW ===")
display(summary_df)
```
